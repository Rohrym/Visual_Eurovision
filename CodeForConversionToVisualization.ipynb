# Eurovision audiofeatures conversion for visualizations

```python
import pandas as pd

def column_normalizer(df, column_name):
    df[column_name] = (df[column_name] - df[column_name].min()) / (df[column_name].max() - df[column_name].min())
    return df[column_name]
```

```python
df = pd.read_json('all_songs.json')
column_normalizer(df,'tempo')
column_normalizer(df,'loudness')
column_normalizer(df,'song_popularity')
df
```

```python
df.to_csv(r'audiofeaturesall.csv')
```

## Audiofeatures as measured per year

```python
af_year = df.groupby('song_year')['danceability','energy', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'loudness', 'liveness', 'valence'].mean()
#af_year
```

```python
af_year.to_csv(r'audiofeaturesyearly.csv')
```
